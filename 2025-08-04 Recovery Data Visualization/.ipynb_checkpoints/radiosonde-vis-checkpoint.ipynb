```python
import pandas as pd
import matplotlib.pyplot as plt
from metpy.plots import SkewT
from metpy.units import units
from metpy.calc import wind_components
import numpy as np

# Load data
df = pd.read_csv("2025080312-72426.csv")

# Extract and convert data
pressure = df['pressure_hPa'].values * units.hPa
temperature = df['temperature_C'].values * units.degC
dewpoint = df['dew point temperature_C'].values * units.degC
wind_speed = df['wind speed_m/s'].values * units('m/s')
wind_dir = df['wind direction_degree'].values * units.degrees

# Convert wind speed to knots
wind_speed_kts = wind_speed.to('knots')

# Compute u and v wind components
u, v = wind_components(wind_speed_kts, wind_dir)

# Filter: only use pressure >= 100 hPa
mask = pressure >= 100 * units.hPa

pressure = pressure[mask]
temperature = temperature[mask]
dewpoint = dewpoint[mask]
u = u[mask]
v = v[mask]

# Choose pressure levels to plot wind barbs (every 25 hPa)
levels = np.arange(1000, 100 - 1, -25) * units.hPa
idx = [np.abs(pressure - lvl).argmin() for lvl in levels if np.any(pressure >= lvl)]

# Create high-resolution Skew-T plot (poster quality)
fig = plt.figure(figsize=(12, 12), dpi=300)  # Large size + high DPI
skew = SkewT(fig, rotation=45)

# Plot temperature and dew point
skew.plot(pressure, temperature, 'r')
skew.plot(pressure, dewpoint, 'g')

# Plot thinned wind barbs at selected levels
skew.plot_barbs(pressure[idx], u[idx], v[idx])

# Add dry/moist adiabats and mixing lines
skew.plot_dry_adiabats()
skew.plot_moist_adiabats()
skew.plot_mixing_lines()

# Title and layout
plt.title("National Weather Service Wilmington, Ohio: 2025-08-03 12Z Radiosonde Data as Skew-T Log-P\nCollected by Graw DFM-17, S/N 24348-082199, recovered by KE8ZDI and W3WAM at 39.4012°N, 83.4738°W")
plt.xlabel('°C')
plt.ylabel('Pressure (mbar)')
plt.tight_layout()

# Save high-resolution version (optional)
plt.savefig("2025-08-03-12Z_KILN_Sounding_12in_square.png", dpi=300)
plt.savefig("2025-08-03-12Z_KILN_Sounding_12in_square.svg", dpi=300)

plt.show()
```
